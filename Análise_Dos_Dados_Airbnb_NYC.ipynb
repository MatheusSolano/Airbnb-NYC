### **Q3. Existem valores Outliers?**

Para podermos identificar se existem valores discrepantes(outliers), vamos utilizar algumas mtricas estatsticas e logo em seguida plotar boxplots, e vamos calcular os limites utilizando a regra do IQR Score.

O IQR no  nada mais do que: 

* Subtrair o Terceiro Quartil(75%) pelo Primeiro Quartil(25%).

Vamos dar uma olhada nos dados para idendificarmos o que ser preciso fazer.



```python
#Verificando a distribuio estatstica 
df[['price', 'Preo em R$', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].describe()
```

Podemos notar algumas coisas que chamam nossa ateno, como por exemplo:

* A varivel `minimum_nights` tem como valore maxmo 1250, sendo que o Q3 est em 5.

* A coluna de `Preo em R$`, 75% do mesmo est abaixo de 883.75 reais, e o seu valor mximo  R$ 50.500.

* A varivel `Preo em R$` tem o valor mnimo 0.

* Podemos observar que acima de 30 noites, j so valores extremamente discrepantes. Pois no faz sentido um aluguel com um mnimo de mais de 30 dias para poder alugar.

```python
#Verificando as distribuies
df.hist(bins=15, figsize=(15, 12), color='royalblue')
plt.show()
```

Verificamos os histogramas, conseguimos ver claramente que existem outliers. Para trat-los vamos seguir os seguintes passos:

* Definir o Q1 e Q3 para as variveis que sero limpas.
* Calcular o IQR para as variveis.
* Definir o limite superior e inferior para cortar os outliers.
* Remover os outliers.

#### **Tratando Valores Outliers**

```python
#Identificando os outliers para a variavel price e Preo em R$
q1_price = df['price'].quantile(.25)
q3_price = df['price'].quantile(.75)
IQR_price = q3_price - q1_price
print(f"IQR da varivel price: {IQR_price}")

q1_preco = df['Preo em R$'].quantile(.25)
q3_preco = df['Preo em R$'].quantile(.75)
IQR_preco = q3_preco - q1_preco
print(f"IQR da varivel Preo em R$: {IQR_preco}")

#Definindo os limites
sup_price = q3_price + 1.5 * IQR_price
inf_price = q3_price - 1.5 * IQR_price
print(f"Limite superior de price: {sup_price}")
print(f"Limite inferior de price: {inf_price}")

sup_preco = q3_preco + 1.5 * IQR_preco
inf_preco = q3_preco - 1.5 * IQR_preco
print(f"Limite superior de preco: {sup_preco:.2f}")
print(f"Limite inferior de preco: {inf_preco:.2f}")
```

```python
# verificando o conjunto original
fig, ax = plt.subplots(figsize=(15,3))
df['price'].plot(kind='box', vert=False);
ax.set_title('Dataset Original - price')
plt.show()
print("O dataset possui {} colunas".format(df.shape[0]))
print("{} Entradas acima de 334".format(len(df[df['price'] > 334])))
print("Representam {:.2f}% do dataset".format((len(df[df['price'] > 334]) / df.shape[0])*100))
```

```python
# identificando os outliers para a varivel minimum_nights
q1_minimum_nights = df.minimum_nights.quantile(.25)
q3_minimum_nights = df.minimum_nights.quantile(.75)
IQR_minimum_nights = q3_minimum_nights - q1_minimum_nights
print('IQR da varivel minimum_nights: ', IQR_minimum_nights)

# definindo os limites                                       
sup_minimum_nights = q3_minimum_nights + 1.5 * IQR_minimum_nights
inf_minimum_nights = q1_minimum_nights - 1.5 * IQR_minimum_nights

print('Limite superior de minimum_nights: ', sup_minimum_nights)
print('Limite inferior de minimum_nights: ', inf_minimum_nights)
```

```python
# verificando o conjunto original
fig, ax = plt.subplots(figsize=(15,3))
df["minimum_nights"].plot(kind='box', vert=False);
ax.set_title('Dataset Original - minimum_nights')
plt.show()
print("O dataset possui {} colunas".format(df.shape[0]))
print("{} Entradas acima de 11.0".format(len(df[df["minimum_nights"] > 11.0])))
print("Representam {:.2f}% do dataset".format((len(df[df["minimum_nights"] > 11.0]) / df.shape[0])*100))
```

```python
#Limpando o Dataset
df_clean = df.copy()

df_clean.drop(df_clean[df_clean["price"] > 334].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean["price"] == 0.0].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean["Preo em R$"] > 1686.70].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean["Preo em R$"] == 0.0].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean["minimum_nights"] > 11].index, axis=0, inplace=True)

print('Shape antes da limpeza: ', df.shape)
print('Shape aps a limpeza: ',df_clean.shape)
```

### **Q4. Existe alguma correlao entre as variveis?**

Atravz da correlao podemos descobrir se h algum ligamento ou relao entre duas ou mais variveis. Essa relao pode ser medida, e  funo do coeficiente de correlao estabelecer qual a intensidade dela.

Para identificar as correlaes existentes, vou:
* Criar uma matriz de correlao
* Gerar um *mapa de calor* utilizando a matriz

* Quando olhamos para esse mapa de calor, intuitivamente j deduzimos que as cores mais fortes, tendem a ter uma correlao maior.

#### **Qual  o valor do aluguel mais caro de cada regio? E o mais barato?**

```python
#Mais caro por regio
df_max = (df_clean.groupby('neighbourhood_group')['price'].max()).sort_values(ascending=False)
df_maxr = (df_clean.groupby('neighbourhood_group')['Preo em R$'].max()).sort_values(ascending=False)

print('Em dolar: {}'.format(df_max))
print('\nEm Real: {}'.format(df_maxr))
```

```python
#Mais barato por regio
price = df_clean[df_clean['price'] > 0]
preco = df_clean[df_clean['Preo em R$'] > 0]

df_min = price.groupby('neighbourhood_group')['price'].min()
df_minr = preco.groupby('neighbourhood_group')['Preo em R$'].min()

print('Em dolar: {}'.format(df_min))
print('\nEm Real: {}'.format(df_minr))
```

* Nas duas clulas acima agrupei os dados pelas regies, e a partir da pude usar as funes para poder extrair os valores mais baratos e mais caros  de cada uma das regies.

#### **Qual a soma dos valores do aluguel por regio? E qual o mais caro?**

```python
#Vamos plotar um grfico de barras para melhor vizualizao
cores = {'Manhattan': 'red',
         'Brooklyn': 'darkorange',
         'Queens': 'yellow',
         'Bronx': 'lawngreen',
         'Staten Island':'dodgerblue'}
labels = list(cores.keys())

y = (df_clean.groupby('neighbourhood_group')['price'].sum()).sort_values(ascending=False)

fig = plt.figure(figsize=(8,5))
ax = fig.add_axes([1,1,1,1])

ax.bar(labels, y, color=cores.values(), alpha=0.9)
ax.set_title('Regies mais carass (Ordem decrescente)', fontsize=18, pad=15)
ax.set_xlabel('Regies', fontsize=15)
ax.set_ylabel('Preo Em Dolar', fontsize=15)

ax.set_yticks([0, 500000, 1000000, 1500000, 2000000, 2500000, 3000000])
ax.set_yticklabels(['0', '500.000', '1.000.000', '1.500.000', '2.000.000', '2.500.000', '3.000.000'])

handles = [plt.Rectangle((0,0),1,1, color=cores[label]) for label in labels]
ax.legend(handles, labels)
plt.show()

print(y)
```

 * Observando o grfico, chegamos a concluso que a regio mais cara de Nova York  a de Manhattan, vejamos a seguir o por que.

### **Q6. Conseguimos saber onde esto localizados os imveis com o valor do aluguel mais caro?**

Para finalizar nossa anlise, vou montar um mapa aonde esto localizados os imveis com o valor do aluguel mais caro.

```python
#Conjunto de dados com as coordenadas
colunas = ['price','neighbourhood_group','latitude','longitude']
data_plot = df_clean.loc[:, colunas].groupby('neighbourhood_group').max().reset_index()

#Um comando para desenhar um mapa
mapa = folium.Map()

#Colocar os pontos de latitude e longitude dentro do mapa
for index, location_info in data_plot.iterrows():
    folium.Marker([location_info['latitude'], location_info['longitude']],
                  popup=location_info['neighbourhood_group']).add_to(mapa)
    
mapa
```
